```python
def text_preprocessing(quo):
  quo=re.sub('[^a-zA-Z]',' ',quo)
  quo=quo.lower().split()
  ps=PorterStemmer()
  clean_word=[ps.stem(i) for i in quo if not i in sw]
  sen=' '.join(clean_word)
  return sen
```

```python
data['clean_word']=data["Review"].apply(text_preprocessing)
data.head()
```
